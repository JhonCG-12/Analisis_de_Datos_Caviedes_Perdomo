### 3. Crear el DataFrame

```python
import pandas as pd
import numpy as np
data = {
    'Fecha': pd.to_datetime(['2023-01-10', '2023-01-11', '2023-01-10', '2023-01-12', '2023-01-11',
                          '2023-01-12', '2023-01-13', '2023-01-14', '2023-01-13', '2023-01-15',
                          '2023-01-14', '2023-01-15', '2023-01-16', '2023-01-17', '2023-01-16']),
    'Ciudad': ['Bogotá', 'Medellín', 'Bogotá', 'Cali', 'Medellín', 'Bogotá', 'Cali', 'Medellín',
               'Bogotá', 'Cali', 'Medellín', 'Bogotá', 'Medellín', 'Cali', 'Bogotá'],
    'Producto': ['Laptop', 'Mouse', 'Teclado', 'Laptop', 'Monitor', 'Mouse', 'Teclado', 'Laptop',
                 'Monitor', 'Mouse', 'Teclado', 'Monitor', 'Mouse', 'Laptop', 'Teclado'],
    'Cantidad': [10, 25, 15, 8, 12, 20, 18, 7, 9, 30, 14, 11, 22, 6, 16],
    'Precio_Unitario': [1200, 25, 50, 1350, 300, 26, 55, 1400, 320, 24, 52, 310, 27, 1300, 53],
    'Vendedor_ID': [101, 102, 101, 103, 102, 101, 103, 102, 101, np.nan, 102, 101, 102, 103, np.nan]
}
df_ventas = pd.DataFrame(data)
print('DataFrame creado exitosamente')
```

### 13. Crear una Nueva Columna
Crea una nueva columna llamada `Ingreso_Total` que sea el resultado de multiplicar `Cantidad` por `Precio_Unitario`.

```python
# Tu código aquí
df_ventas['Ingreso_Total'] = df_ventas['Cantidad'] * df_ventas['Precio_Unitario']
df_ventas

```

### 14. Manejo de Valores Nulos
Rellena los valores nulos en la columna `Vendedor_ID` con el valor `0`.

```python
# Tu código aquí
df_ventas['Vendedor_ID'] = df_ventas['Vendedor_ID'].fillna(0)
df_ventas

```

### 15. Aplicar una Función (`.apply`)
Crea una columna `Categoria_Precio` que contenga 'Alto' si el `Precio_Unitario` es mayor a 500, y 'Bajo' en caso contrario. Usa una función y `.apply()`.

```python
# Tu código aquí
df_ventas['Categoria_Precio'] = df_ventas['Precio_Unitario'].apply(
    lambda x: 'Alto' if x > 500 else 'Bajo'
)
df_ventas
```

### 16. Ordenar el DataFrame
Ordena `df_ventas` por la columna `Ingreso_Total` de forma descendente.

```python
# Tu código aquí
df_ventas.sort_values('Ingreso_Total', ascending=False)
```

### 17. Agrupación Simple (`.groupby`)
Agrupa el DataFrame por `Ciudad` y calcula la suma de `Ingreso_Total` para cada ciudad.

```python
# Tu código aquí
df_ventas.groupby('Ciudad')['Ingreso_Total'].sum()
```

### 18. Agrupación con Múltiples Agregaciones
Agrupa por `Producto` y calcula la cantidad promedio (`mean`) vendida y el ingreso total (`sum`) para cada producto.

```python
# Tu código aquí
df_ventas.groupby('Producto').agg(
    Promedio_Cantidad=('Cantidad', 'mean'),
    Total_Ingresos=('Ingreso_Total', 'sum')
)
```

### 19. Encontrar el Producto más Vendido
Encuentra el nombre del producto con el mayor `Ingreso_Total` acumulado.

```python
# Tu código aquí
df_ventas.groupby('Producto')['Ingreso_Total'].sum().idxmax()

```
